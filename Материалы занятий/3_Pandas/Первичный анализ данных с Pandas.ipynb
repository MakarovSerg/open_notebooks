Датафреймы можно индексировать как по названию столбца или строки, так и по порядковому номеру. Для индексации **по названию** используется метод **`loc`**, **по номеру** — **`iloc`**.

В первом случае мы говорим _«передай нам значения для id строк от 0 до 5 и для столбцов от State до Area code»_, а во втором — _«передай нам значения первых пяти строк в первых трёх столбцах»_. 

В случае `iloc` срез работает как обычно, однако в случае `loc` учитываются и начало, и конец среза. 

````python
tmp = pd.DataFrame({'a':[1,2,3],'b':[3,4,5]}, index=['ind_1','ind_2','ind_3'])
tmp
````

````python
tmp.iloc[1]
````

````python
tmp.loc['ind_1']
````

````python
df.loc[0:5, "State":"Area code"]
````

````python
df.loc[:,  "State":"Area code"]
````

````python
df.iloc[0:5, 0:3]
````

### Группировка данных

В общем случае группировка данных в Pandas выглядит следующим образом:

```
df.groupby(by=grouping_columns)[columns_to_show].function()
```

1. К датафрейму применяется метод **`groupby`**, который разделяет данные по `grouping_columns` – признаку или набору признаков.
3. Индексируем по нужным нам столбцам (`columns_to_show`). 
2. К полученным группам применяется агрегирующая функция или несколько функций.

````python
# Группировка по одному полю
df.groupby('Churn').count()
````

````python
df.groupby('Churn')[['State']].count()
````

````python
# Переименуем столбец
df.groupby('Churn')[['State']].count().rename(columns={'State':'Number of abonents'})
````

````python
df.groupby('State')[['Total day charge']].sum()
````

````python
# Группировка по нескольким полям
df.groupby(['Churn', 'International plan'])[['Total day minutes']].count()
````

````python
# агрегирующие функции
'''
count() - количество значений в группе
first(), last() - первое и последнее значение в группе
mean(), median() - среднее арифметическое и медиана
min(), max() - минимальное и максимальное значения
std(), var() - стандартное отклонение и дисперсия
mad() - среднее абсолютное отклонение 
sum() - сумма
prod() - произведение
describe() - расчет всех описательных статистик
'''
````

````python
columns_to_show = ["Total day minutes", "Total eve minutes", "Total night minutes"]

df.groupby(["Churn"])[columns_to_show].describe()
````

Можно применять несколько агрегирующиъ функций в одной группировке

````python
columns_to_show = ["Total day minutes", "Total eve minutes", "Total night minutes"]

df.groupby(["Churn"])[columns_to_show].agg(['min', 'max', 'count'])
````

Можно так же к каждому столбцу применить свою агрегирующую функцию

````python
df.groupby(["Churn"]).agg({"Total day minutes": 'sum', 'Total eve minutes':['min','max','mean']})
````

````python
# Не забывайте присваивать новой переменной результат группировки!
tmp = df.groupby(["Churn"]).agg({"Total day minutes": 'sum', 'Total eve minutes':['min','max','mean']})
tmp
````

### Пропуски в данных

**Поиск пропусков**

Иногда в наборах данных возможно пропуски значений. Проверим, есть ли такие в нашем наборе можно несколькими способами.

Можно сравнить количесво значений по столбцам через метод `describe()` или же с использованием специального метода `isna()`

````python
df.describe()
````

````python
# Проверяем, в каких столбцах есть пропуски
df.isna().any()
````

````python
# маска значений по одному столбцу
df['Number vmail messages'].isna() #isnull
````

````python
# фильтрация датафрейма по пустым значениям в столбце
df[df['Number vmail messages'].isna()]
````

````python
df.loc[:,:]
````

````python
# отфильтровать весь датафрейм не получится - получите все значения NaN
df[df.isna()]
````

Пропуски можно **удалить** из данных либо чем-то **заполнить**, например, нулем, или каким-то другим значением. Чем заполнять - зависит от конкретного набора даннх и задачи

````python
# удаление строк, в которых есть хть один пропуск
df = df.dropna() #inplace=True
````

````python
tmp = pd.DataFrame({'a':[1,2,3], 'b':[3, np.nan, 5]})
tmp.fillna(0, inplace=True)
tmp
````

````python
# заполнение пропусков нулем
df.fillna(0) #inplace=True
df
````
